**This is highly imbalanced dataset**

**0-> Legit Transcation**


**1-> Fraud Transcation**

```python
# Separate the legit and fraudelent

legit =data[data.Class==0]
fraud =data[data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

**Undersampling**

```python
# Building sample dataset containing simmilar distribution of legit and fraudelnt transcation
```

```python
legit_sample= legit.sample(n=364)
```

```python
print(legit_sample.shape)
```

**concatenate two dataset**

```python
new_data=pd.concat([legit_sample,fraud],axis=0)
```
